```python
def display_top5_images(path_To_class,className):
  fig, axes = plt.subplots(1, 5, figsize=(20, 20))
  axes = axes.flatten()
  for i, ax in enumerate(axes):
    img_path = os.path.join(path_To_class+"/"+className, os.listdir(path_To_class+"/"+className)[i])
    img = plt.imread(img_path)
    ax.imshow(img)
    ax.set_title(className)
    ax.axis('off')
display_top5_images(train_dir,"fire")
display_top5_images(train_dir,"nofire")
```
